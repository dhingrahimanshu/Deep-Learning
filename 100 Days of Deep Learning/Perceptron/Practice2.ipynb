```python
from sklearn.datasets import make_classification
import numpy as np
x, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=20,class_sep=3)
```

```python
|import matplotlib.pyplot as plt

plt.scatter(x[: , 0] , x[: , 1]  , c=y , cmap = 'winter' , s = 100)
```

```python
def perceptron(x , y):
    lr = 0.01
    w1 = w2 = b = 1
    
    for i in range(1000):
        for j in range(x.shape[0]):
            fxi = w1*x[j][0] + w2*x[j][1]+b
            
            if -(fxi*y[j]) <0:
                w1 += lr*y[j]*x[j][0]
                w2 += lr*y[j]*x[j][1]
                b += lr*y[j]
                
    return b , (w1 , w2)
```

```python
intercept , coef = perceptron(x,y)
```

```python
print(intercept , coef)
```

```python
y0 = -(coef[0]/coef[1])
c = -(intercept/coef[1])

line_input = np.linspace(-3 , 3 , 100)
line_output = y0*line_input + c
```

```python
plt.plot(line_input , line_output , color = 'red')

plt.scatter(x[: , 0] , x[: , 1] , c = y , cmap = 'winter' , s = 100)
```
