```python
from sklearn.datasets import make_classification

x ,y = make_classification(n_samples = 1000 , n_features=2 , n_informative = 1 , n_redundant =0 , n_classes = 2 , n_clusters_per_class = 1 , class_sep = 2 , random_state = 13 )
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.scatter(x[: , 0] , x[: , 1] , c= y , cmap = 'winter' , s = 10)
```

```python
def perceptron(x , y):
    weights = np.ones(x.shape[1]).reshape(1 , -1)
    b = 1
    lr = 0.001
    for i in range(5000):
        for j in range(x.shape[0]):
            z = -np.dot(weights , x[j]) +b
            # print(z)
            if (z*y[j])<0:
                weights += lr*x[j]*y[j]
                b += lr*y[j]
    
    
    return b, weights
```

```python
intercept , coef = perceptron( x, y)
```

```python
coef = coef.reshape(-1 , 1)
```

```python
y0 = -(coef[0]/coef[1])
c = -(intercept/coef[1])
```

```python
line_input = np.linspace(-10 , 10, 1000)

line_output = y0*line_input +c
```

```python
plt.plot(line_input , line_output , color = 'red')

plt.scatter(x[: , 0] , x[: , 1] , c = y , cmap = 'winter' , s = 10)
plt.ylim(-3 , 3)
plt.xlim(-3 , 3)
```
