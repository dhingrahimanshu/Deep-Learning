```python
from sklearn.datasets import make_classification
```

```python
x , y = make_classification(n_samples = 500 , n_features = 2 , n_informative = 1, n_redundant = 0 , n_classes = 2, n_clusters_per_class = 1 , class_sep =2 , random_state = 1)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x[: , 0] , x[: , 1] , c=y , cmap  = 'winter' , s = 10)
```

```python
import numpy as np
```

```python
def perceptron(x , y , epochs = 1000):
    b = 1
    weights = np.ones(x.shape[1])
    lr = 0.001
   
    for i in range(5000):
        for j in range(x.shape[0]):
            z = -np.dot(weights , x[j]) +b
            if (z*y[j])<0:
                weights = weights + lr*x[j]*y[j]
                b = b + lr*y[j]
        
      
            
    return b , weights
```

```python
b,coef = perceptron(x ,y)
```

```python
print(b , coef)
```

```python
y0 = -(coef[0]/coef[1])
c = -(b/coef[1])
```

```python
input_line = np.linspace(-3 , 7 , 100)
output_line = y0*input_line +c

plt.plot(input_line , output_line , color = 'red')
plt.scatter(x[: , 0] , x[: , 1] , c = y , cmap = 'winter' , s= 10)
plt.ylim(-3 , 3)
plt.xlim(-3 , 3)
```
